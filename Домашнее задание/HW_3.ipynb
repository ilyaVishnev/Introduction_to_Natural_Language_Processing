### Задание 1. 
Spacy tagger. Популярные NER, персоны, организации

```python
 # Объединяем все слова в 1 список
def words_union(df,colum='tweet'):
    words_list = []
    for line in df[colum]:
        line_words = str(line).split()
        words_list.extend(line_words)
    words = ' '.join(words_list)
    return words
```

```python
words = words_union(combine_df)
max_word_limit_spacy = 1000000
nlp = spacy.load('en_core_web_sm')#en_core_web_md.load() 
doc = nlp(words[:max_word_limit_spacy])
tokens_list = [(ent.text, ent.label_) for ent in doc.ents]
tokens_list
```

```python
# TOP-20 популярных NER
collections.Counter(tokens_list).most_common(20)
```

```python
# TOP-20 популярных личностей
person_tokens_list = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ == 'PERSON']
collections.Counter(person_tokens_list).most_common(20)
```

```python
# TOP-20 популярных организаций
org_tokens_list = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ == 'ORG']
collections.Counter(org_tokens_list).most_common(20)
```

```python
# сколько уникальных NER найдено с помощью spacy?
len(collections.Counter(tokens_list))
```

### Задание 2.

```python
nltk.download(['averaged_perceptron_tagger', 'maxent_ne_chunker','words','svgling'])
```

```python
df = pd.read_csv("lenta-ru-news.csv")
words=words_union(df,'text')
```

```python
# words = words_union(combine_df)
max_word_limit_spacy = 1000000 #if len(words) > 1000000 else len(words.split())
tokens_list =[(' '.join(c[0] for c in chunk), chunk.label() ) for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(words[:max_word_limit_spacy]))) if hasattr(chunk, 'label') ]
```

```python
tokens_list
```

```python
# TOP-20 популярных NER
collections.Counter(tokens_list).most_common(20)
```

```python
# TOP-20 популярных личностей
person_tokens_list = [(ent[0], ent[1]) for ent in tokens_list if ent[1] == 'PERSON']
collections.Counter(person_tokens_list).most_common(20)
```

```python
# TOP-20 популярных организаций
org_tokens_list = [(ent[0], ent[1]) for ent in tokens_list if ent[1] == 'ORGANIZATION']
collections.Counter(org_tokens_list).most_common(20)
```

```python
# сколько уникальных NER найдено с помощью spacy?
len(collections.Counter(tokens_list))
```
